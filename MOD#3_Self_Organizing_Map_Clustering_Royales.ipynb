```python
import pandas as pd
import numpy as np
from minisom import MiniSom
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('E:heart_cleveland_upload.csv')

# Attribute Selection
selected_attributes = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
target_attribute = 'condition'

# Set the class (target) and attributes (predictors)
X = data[selected_attributes].values
y = data[target_attribute].values

# Normalize the data
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Define the SOM parameters
map_width = 10
map_height = 10
input_len = X.shape[1]
learning_rate = 0.5
sigma = 1.0
num_epochs = 100

# Create and train the SOM
som = MiniSom(map_width, map_height, input_len, sigma=sigma, learning_rate=learning_rate)
som.random_weights_init(X)
som.train_random(X, num_epochs)

# Visualize the SOM
plt.figure(figsize=(map_width, map_height))
plt.pcolor(som.distance_map().T, cmap='bone_r')
plt.colorbar()

# Mark the class assignments on the SOM
target_labels = np.unique(y)
target_colors = ['r', 'g', 'b', 'c', 'm']
for idx, sample in enumerate(X):
    w = som.winner(sample)
    plt.plot(w[0] + 0.5, w[1] + 0.5, marker='o', markersize=8, markeredgecolor=target_colors[y[idx]], markerfacecolor='None', markeredgewidth=2)
plt.title('Self-Organizing Map')
plt.show()
```
